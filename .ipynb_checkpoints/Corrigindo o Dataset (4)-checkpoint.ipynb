# Verificando Outliers

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Dados/6. bigmart.csv')
df.head()
```

```python
# configurar a visualização
#%matplotlib inline
%config InlineBackend.figure_format = 'svg'
#configurando parâmetros das fontes
plt.rcParams['font.family'] = 'sans-serif'
```

```python
df.describe()
```

```python
# Plotando varios graficos em uma grade
Linhas = 1
Colunas = 4

# Ajustando a figura
plt.figure(figsize=(15,4))

# Adicionando eixos
ax = plt.axes()

# Primeiro Grafico
plt.subplot(Linhas,Colunas,1)
sns.boxplot(df['Item_Weight'])
plt.title('Grafico 1')

# Segundo Grafico
plt.subplot(Linhas,Colunas,2)
sns.boxplot(df['Item_Visibility'])
plt.title('Grafico 2')


# Terceiro Grafico
plt.subplot(Linhas,Colunas,3)
sns.boxplot(df['Item_MRP'])
plt.title('Grafico 3')


# Quarto Grafico
plt.subplot(Linhas,Colunas,4)
sns.boxplot(df['Item_Outlet_Sales'])
plt.title('Grafico 4')

plt.text(
        x = -25000,
        y = -.65, # Coordenada para o Titulo
        s = 'Gráficos Separados', # Texto
        fontsize=16, # Tamanho do Texto
        color='Black', # Cor do Texto
        weight='bold' #
);
```

```python
# Plotando varios graficos em uma grade
Linhas = 1
Colunas = 3

# Ajustando a figura
plt.figure(figsize=(17,4))

# Primeiro Grafico
plt.subplot(Linhas,Colunas,1)
sns.scatterplot(y=df['Item_Outlet_Sales'],x=df['Item_Weight'])
plt.title('Grafico 1')

# Segundo Grafico
plt.subplot(Linhas,Colunas,2)
sns.scatterplot(y=df['Item_Outlet_Sales'],x=df['Item_Visibility'])
plt.title('Grafico 2')


# Terceiro Grafico
plt.subplot(Linhas,Colunas,3)
sns.scatterplot(y=df['Item_Outlet_Sales'],x=df['Item_MRP'])
plt.title('Grafico 3')

plt.text(
        x = -250,
        y = 16000, # Coordenada para o Titulo
        s = 'Gráficos Separados', # Texto
        fontsize=16, # Tamanho do Texto
        color='Black', # Cor do Texto
        weight='bold' #
);
```

# Capping outliers

```python
df['Item_Outlet_Sales'].values[df['Item_Outlet_Sales'].values>10000]=10000
```

```python
df[df['Item_Outlet_Sales']>10000]
```

```python
df=df[~(df['Item_Outlet_Sales']>8000)]
```

```python
# Plotando varios graficos em uma grade
Linhas = 1
Colunas = 4

# Ajustando a figura
plt.figure(figsize=(15,4))

# Primeiro Grafico
plt.subplot(Linhas,Colunas,1)
sns.boxplot(df['Item_Weight'])
plt.title('Grafico 1')

# Segundo Grafico
plt.subplot(Linhas,Colunas,2)
sns.boxplot(df['Item_Visibility'])
plt.title('Grafico 2')


# Terceiro Grafico
plt.subplot(Linhas,Colunas,3)
sns.boxplot(df['Item_MRP'])
plt.title('Grafico 3')


# Quarto Grafico
plt.subplot(Linhas,Colunas,4)
sns.boxplot(df['Item_Outlet_Sales'])
plt.title('Grafico 4')

plt.text(
        x = -25000,
        y = -.65, # Coordenada para o Titulo
        s = 'Gráficos Separados', # Texto
        fontsize=16, # Tamanho do Texto
        color='Black', # Cor do Texto
        weight='bold' #
);
```
