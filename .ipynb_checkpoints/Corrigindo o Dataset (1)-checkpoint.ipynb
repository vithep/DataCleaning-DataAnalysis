# Verificando o Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Dados/6. bigmart.csv')
```

```python
data.shape
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
x=data.isnull().sum()
y=(data.isnull().sum()/data.shape[0])*100
z={'Number of missing values':x,'Percentage of missing values':y}
df=pd.DataFrame(z,columns=['Number of missing values','Percentage of missing values'])
df.sort_values(by='Percentage of missing values',ascending=False)
```

```python
data.info()
```

```python
data['Item_Weight']=data['Item_Weight'].fillna(data['Item_Weight'].mean())
data['Outlet_Size']=data['Outlet_Size'].fillna(data['Outlet_Size'].mode()[0])
```

```python
data.isnull().sum()
```

```python
# configurar a visualização
#%matplotlib inline
%config InlineBackend.figure_format = 'svg'
#configurando parâmetros das fontes
plt.rcParams['font.family'] = 'sans-serif'
```

```python
# Plotando varios graficos em uma grade
Linhas = 1
Colunas = 3

# Ajustando a figura
plt.figure(figsize=(15,4), 
           facecolor='gray') # Ajustando a cor

# Adicionando eixos
ax = plt.axes()
ax.set_facecolor('white') #Cor de fundo do gráfico

# Primeiro Grafico
plt.subplot(Linhas,Colunas,1)
sns.boxplot(data['Item_Weight'])
plt.title('Grafico 1')

# Segundo Grafico
plt.subplot(Linhas,Colunas,2)
sns.boxplot(data['Item_Visibility'])
plt.title('Grafico 2')


# Terceiro Grafico
plt.subplot(Linhas,Colunas,3)
sns.boxplot(data['Item_MRP'])
plt.title('Grafico 3')

plt.text(
        x = -250,
        y = -0.65, # Coordenada para o Titulo
        s = 'Gráficos Separados', # Texto
        fontsize=16, # Tamanho do Texto
        fontfamily='fantasy', # Tipo da Fonte
        color='Black', # Cor do Texto
        weight='bold' #
);
```
