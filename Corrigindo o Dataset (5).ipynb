# Imputando valores ausentes

```python
import pandas as pd
```

```python
pd.set_option("max_rows",89)
pd.set_option("max_columns",89)
```

```python
dados = pd.read_csv('Dados/fifa.csv')
dados.head()
```

```python
dados.columns
```

```python
# Por Linha
dados.isnull().sum(axis=1).sort_values(ascending=False)
```

```python
# Verificando se tem alguma linha sem valor
dados.isnull().all(axis=1).sum()
```

```python
# Verificando a taxa de valores nulos
faltantes_l = dados.isnull().sum(axis=1)
porcentagem_l = (dados.isnull().sum(axis=1)/dados.shape[1])*100
verf = {'Número de valores ausentes': faltantes_l, 'Porcentagem de ausentes':porcentagem_l}
tabela = pd.DataFrame(verf,columns=['Número de valores ausentes', 'Porcentagem de ausentes'])
tabela.sort_values(by='Porcentagem de ausentes',ascending=False)
```

```python
# Temos linhas com mais de 50 valores nulos?
dados[dados.isnull().sum(axis=1)>50]
```

```python
# Tamanho antes da alteração
dados.shape[0]
```

```python
# Tamanho depois da alteração
dados = dados[dados.isnull().sum(axis=1)<=50]
dados.shape[0]
```

```python
# Por coluna
dados.isnull().sum().sort_values(ascending=False)
```

```python
# Verificando a taxa de valores nulos
faltantes = dados.isnull().sum()
porcentagem = (dados.isnull().sum()/dados.shape[0])*100
verf = {'Número de valores ausentes': faltantes, 'Porcentagem de ausentes':porcentagem}
tabela = pd.DataFrame(verf,columns=['Número de valores ausentes', 'Porcentagem de ausentes'])
tabela.sort_values(by='Porcentagem de ausentes',ascending=False)
```

```python
# A coluna 'Loaned From' apresenta muito valores nulos
dados=dados.drop(['Loaned From'],axis=1)
```

```python
dados.columns
```

```python
dados.shape
```

```python
dados.dtypes[dados.isnull().any()]
```

```python
# Preenchendo os dados faltantes
#---
dados['Jersey Number'].fillna('NA',inplace=True)
#---
dados['Club']=dados['Club'].fillna(dados['Club'].mode()[0])
dados['Position']=dados['Position'].fillna(dados['Position'].mode()[0])
dados['Joined']=dados['Joined'].fillna(dados['Joined'].mode()[0])
dados['Contract Valid Until']=dados['Contract Valid Until'].fillna(dados['Contract Valid Until'].mode()[0])
dados['Release Clause']=dados['Release Clause'].fillna(dados['Release Clause'].mode()[0])
#---
dados['RB'].fillna(0,inplace=True)
dados['RCB'].fillna(0,inplace=True)
dados['CB'].fillna(0,inplace=True)
dados['LCB'].fillna(0,inplace=True)
dados['LB'].fillna(0,inplace=True)
dados['RWB'].fillna(0,inplace=True)
dados['RDM'].fillna(0,inplace=True)
dados['CDM'].fillna(0,inplace=True)
dados['LDM'].fillna(0,inplace=True)
dados['LWB'].fillna(0,inplace=True)
dados['RM'].fillna(0,inplace=True)
dados['RCM'].fillna(0,inplace=True)
dados['CM'].fillna(0,inplace=True)
dados['LCM'].fillna(0,inplace=True)
dados['LM'].fillna(0,inplace=True)
dados['RAM'].fillna(0,inplace=True)
dados['CAM'].fillna(0,inplace=True)
dados['LAM'].fillna(0,inplace=True)
dados['RW'].fillna(0,inplace=True)
dados['RF'].fillna(0,inplace=True)
dados['CF'].fillna(0,inplace=True)
dados['LF'].fillna(0,inplace=True)
dados['LW'].fillna(0,inplace=True)
dados['RS'].fillna(0,inplace=True)
dados['ST'].fillna(0,inplace=True)
dados['LS'].fillna(0,inplace=True)
```
